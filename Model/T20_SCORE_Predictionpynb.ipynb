```python
import numpy as np
import pandas as pd
```

```python
deliveries = pd.read_csv('deliverywise_data.csv')
```

```python
deliveries.shape
```

```python
deliveries.sample(10)
```

```python
deliveries.columns
```

```python
deliveries = deliveries[['match_id', 'innings_number', 'batting_team', 'bowling_team',
       'over_number', 'ball_number', 'total_runs', 'player_dismissed']]
```

```python
deliveries.shape
```

```python
deliveries['batting_team'].unique()
```

```python
teams = ['Australia', 'New Zealand', 'England', 'South Africa',
       'West Indies', 'Sri Lanka', 'Pakistan', 'India',
       'Bangladesh', 'Afghanistan','Scotland', 'Canada', 'Zimbabwe',  'Ireland',
       'Netherlands', 'Nepal',
       'United Arab Emirates',  'Oman',
       'Namibia', 'Italy' ]
```

```python
deliveries = deliveries[deliveries['batting_team'].isin(teams)]
deliveries = deliveries[deliveries['bowling_team'].isin(teams)]
```

```python
deliveries = deliveries[deliveries['innings_number'].isin([1])]
```

```python
deliveries.shape
```

```python
deliveries.sample(5)
```

```python
deliveries['current_score'] = deliveries.groupby('match_id')['total_runs'].cumsum()
```

```python
deliveries.head()
```

```python
deliveries['balls_bowled'] =(deliveries['over_number'].astype('int')*6) + deliveries['ball_number'].astype('int')
deliveries['balls_left'] = 120 - (deliveries['balls_bowled'])
```

```python
deliveries['balls_left'] = deliveries['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
deliveries['player_dismissed'] = deliveries['player_dismissed'].notna().astype(int)
```

```python
deliveries.sample(10)
```

```python
deliveries['wicket_fall'] = deliveries.groupby('match_id')['player_dismissed'].cumsum()
deliveries['wicket_left'] = (10 - deliveries['wicket_fall'])
```

```python
deliveries
```

```python
deliveries['CRR'] = (deliveries['current_score']*6)/deliveries['balls_bowled']
```

```python
deliveries.sample(5)
```

```python
# Ensure correct sort
deliveries = deliveries.sort_values(['match_id','over_number','ball_number']).copy()

# Compute rolling sum of last 30 balls per match (5 overs)
deliveries['last_5'] = (
    deliveries.groupby('match_id')['total_runs']
    .rolling(window=30, min_periods=1)  # last 30 balls (5 overs)
    .sum()
    .reset_index(level=0, drop=True)
)

# Set last_5 = 0 if over_number < 5
deliveries.loc[deliveries['over_number'] < 5, 'last_5'] = 0

```

```python
deliveries
```

```python
deliveries['total_score'] = deliveries.groupby('match_id')['total_runs'].transform('sum')
```

```python
deliveries
```

```python
deliveries.describe()
```

```python
deliveries.drop(columns=['batting_team','bowling_team']).corr()
```

```python
deliveries.isnull().sum()
```

```python
deliveries = deliveries[(deliveries['last_5'])!=0]
deliveries.dropna(inplace=True)
```

```python
deliveries
```

```python
deliveries.describe()
```

```python
deliveries.shape
```

```python
deliveries = deliveries[['batting_team', 'bowling_team','current_score','balls_left','wicket_left', 'CRR', 'last_5', 'total_score']]
```

```python
deliveries
```

```python
final_df = deliveries.sample(deliveries.shape[0])
```

```python
final_df.drop(columns=['batting_team', 'bowling_team']).corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import Ridge
```

```python
x = final_df.drop(columns=['total_score'])
y = final_df['total_score']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team']),
    ('scaling', StandardScaler(), ['current_score', 'balls_left', 'CRR', 'last_5', 'wicket_left'])],
                        remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', XGBRegressor(learning_rate= 0.1, max_depth= 7, n_estimators= 500, subsample=0.8))
])
```

```python
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
pipe.predict(pd.DataFrame([['India','Australia',51,60,6,5.1,1]], columns=['batting_team','bowling_team','current_score','balls_left','wicket_left','CRR','last_5']))
```

```python
param_grid = {
    'step2__n_estimators': [100, 200, 500],
    'step2__max_depth': [3, 5, 7],
    'step2__learning_rate': [0.01, 0.05, 0.1],
    'step2__subsample': [0.7, 0.8, 1]
}


grid = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring='r2',     # maximize R2
    cv=5,
    n_jobs=-1
)

grid.fit(x_train, y_train)S
```
